```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
raw_data = pd.read_csv('movies.csv')
raw_data.head(3)
```

```python
raw_data.describe()
```

```python
raw_data.info()
```

```python
raw_data['budget'] = raw_data['budget'].fillna(0)
raw_data['gross'] = raw_data['gross'].fillna(0)
```

```python
raw_data['budget'] = raw_data['budget'].astype('int64')
raw_data['gross'] = raw_data['gross'].astype('int64')
raw_data.head()
```

```python
raw_data['runtime'] = raw_data['runtime'].fillna(0)
raw_data['score'] = raw_data['score'].fillna(0)
raw_data['votes'] = raw_data['votes'].fillna(0)
```

```python
raw_data['runtime'] = raw_data['runtime'].astype('int64')
raw_data['score'] = raw_data['score'].astype('int64')
raw_data['year'] = raw_data['year'].astype('int64')
raw_data['votes'] = raw_data['votes'].astype('int64')
```

```python
raw_data.info()
```

```python
raw_data.head()
```

```python
raw_data['releasedCorrect'] = raw_data['released'].str.extract(r'(\d{4})')
raw_data['releasedCorrect'].head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
raw_data.sort_values(by = ['gross'], inplace = False, ascending=False)
raw_data.head()
```

```python
raw_data.head()
```

```python
sns.regplot(data=raw_data, x='budget',y='gross',scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
numeric_data = raw_data.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr(method='pearson')
print(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
df_numerized = raw_data

for col_name in df_numerized.columns:
  if df_numerized[col_name].dtype == 'object':
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
corr_matrixx = df_numerized.corr()
corr_matrixx
```

```python
sns.heatmap(corr_matrixx, annot=True)
plt.show()
```

```python
corr_mattrix = df_numerized.corr()
corr_unstack = corr_mattrix.unstack()
corr_unstack
```

```python
sorted = corr_unstack.sort_values()
sorted
```

```python
high_correlation = sorted[sorted>0.5]
high_correlation
```
